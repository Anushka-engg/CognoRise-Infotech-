```python
df = pd.DataFrame(data)
# print(df)
gp = data.groupby("job_title").agg({"salary_in_usd":"count"})
print(gp)
```

```python
print(pd.melt(df, id_vars = ["job_title"], value_vars =["experience_level","company_size"], var_name = "experience_level", value_name = "values"))
```

```python
a=pd.melt(df, id_vars = ["job_title"], value_vars =["experience_level","company_size"], var_name = "experience_level", value_name = "values")
plt.figure(figsize=(10, 7))
plt.hist(a, orientation = 'horizontal' )
plt.savefig("hist.png", transparent = True)
```

```python
plt.stem(df["experience_level"], df["salary_in_usd"])
plt.savefig("stem.png")
plt.show()
```

```python
plt.plot(df["salary_in_usd"], df["work_year"], label = "2020")
plt.plot(df["salary_in_usd"], df["work_year"], label = "2021")
plt.plot(df["salary_in_usd"], df["work_year"], label = "2022")
#plt.legend()
plt.show()
plt.savefig("plot.png", transparent = True)
```

```python
grouped_by = df.groupby("job_title")["salary_in_usd"].sum()
#print(grouped_by)
plt.pie(grouped_by.values)
plt.show()
plt.savefig("pie.png", transparent = True)
```

```python
grouped_by = df.groupby(["job_title","experience_level"])["salary_in_usd"].sum()
print(grouped_by)
plt.plot(grouped_by.values)
plt.show()
plt.savefig("plot.png" , transparent = True)
```

```python
gp1 = df.groupby(["company_location","employee_residence"])["remote_ratio"].count()
print(gp1)
plt.pie(gp1)
plt.show()
plt.savefig("pie.png", transparent = True)
```

```python
gp2 = df.groupby("job_title").agg({"work_year":"count","experience_level":"count"})
print(gp2)
sns.histplot(gp2, kde = True)
```

```python
b = (pd.melt(df,id_vars = ["job_title"],value_vars = ["experience_level"]))
print(b)
plt.xlabel("job_title")
plt.hist(b)
plt.savefig('hist.png', transparent = True)
```
